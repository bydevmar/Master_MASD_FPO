```python
import matplotlib.pyplot as plt
import numpy as np

# En supposant que l'axe des x varie de 0 à 25 et l'axe des y de 0 à 45
x = np.linspace(0, 25, 30)  # Réduction du nombre de points de 100 à 30

# Création de trois lignes avec des pentes différentes
y1 = 0.5 * x
y2 = 1.5 * x
y3 = 2.5 * x

plt.scatter(x, y1, marker='D', color='blue')  # 'D' pour un marqueur en forme de diamant
plt.scatter(x, y2, marker='o', color='green')  # 'o' pour un marqueur en forme de point
plt.scatter(x, y3, marker='*', color='red')  # '*' pour un marqueur en forme d'étoile

plt.xlabel('Étiquette de l\'axe des x')
plt.ylabel('Étiquette de l\'axe des y')
plt.title('Graphique de dispersion')

plt.show()
```
